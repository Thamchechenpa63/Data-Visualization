### The Boxplot:

The boxplot is a visual representation of the five-number summary. The components of the five-number summary are the  minimum, maximum, median, 25th percentile, and 75th percentile values of the dataset. We can also spot the mean in the boxplot, if we want. It is a great summary visual representation of the dataset. 

We see from the boxplot: 

- The range of the datasets or total spread.

- The interquantile range or mid-spread.

- The median value of the dataset or center.

- The skewness of the dataset.

```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### The side-by-side boxplots:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))
    
sns.boxplot(x = Air2008.ArrDelay.head(25).dropna(),
                 ax = ax1, vert = True, fliersize = 7, color = 'red')
sns.boxplot(x = Air2008.DepDelay.head(25).dropna(), 
                 ax = ax2, vert = True, fliersize = 7, color = 'purple')

ax1.set_title('Arrival Delay (in minutes)')
ax2.set_title('Departure Delay (in minutes)');
```

### The side-by-side multiple boxplots:

```python
df = Air2008[['DepDelay', 'DayOfWeek', 'ArrDelay']].head(50)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.boxplot(x = "DayOfWeek", y = "DepDelay", data = df, ax = ax1)
sns.boxplot(x = "DayOfWeek", y = "ArrDelay", data = df, ax = ax2)

ax1.set_title('Departure Delay (in minutes)')
ax2.set_title('Arrival Delay (in minutes)');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.boxplot(x = "pclass", y = "fare", hue = "survived", data = titanic, palette="Set3", ax = ax1)
sns.boxplot(x = "pclass", y = "age", hue = "survived", data = titanic, palette="Set3", ax = ax2)

ax1.set_title('Fare')
ax2.set_title('Pclass');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.boxplot(x = "fare", y = "pclass", hue = "survived", data = titanic.head(50), 
            palette="Set3", orient = 'h', ax = ax1, width = .5, whis = 1)
sns.boxplot(x = "age", y = "pclass", hue = "survived", data = titanic.head(25), 
            palette="Set3", orient = 'h', ax = ax2, width = .5, whis = 1)

ax1.set_title('Pclass')
ax2.set_title('Pclass');
```
