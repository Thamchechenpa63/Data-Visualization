```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

### The area plot:

This area plot breaks out the Arrival delay by day of the week. The colors correlate with the day of the week. The counts of the Arrival delays are shown on the vertical axis. Arrival delays in minutes are shown on the horizontal axis.

```python
df = pd.crosstab(Air2008.ArrDelay, Air2008.DayOfWeek).head(10)
df.plot(kind = 'area', figsize = (10, 5), title = 'The Arrival Delay (in minutes)');
```

### The line plot:

This line plot breaks out the Arrival delay by day of the week. The colors correlate with the day of the week. The counts of the Arrival delays are shown on the vertical axis. Arrival delays in minutes are shown on the horizontal axis.

```python
df.plot(kind = 'line', stacked = True, figsize = (10, 5), title = 'The Arrival Delay (in minutes)');
```
