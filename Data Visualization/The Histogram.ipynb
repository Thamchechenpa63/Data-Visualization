### The Histogram:

The histogram is constructed for a quantitative variable. The y-axis of the histogram corresponds to the frequency or relative frequency or relative percentage. The x-axis of the histogram corresponds to the barwidth or binsize. There are no gaps shown between the bars of the histogram. If any bar is far from the others in the histogram, then there are outliers in the dataset. 

We can learn from the histogram:

- The shape of the dataset.

- The skewness of the dataset.

- Whether the data is normally distributed or not.

- The spread of the dataset.

- The center of the dataset.

```python
import os

import pandas as pd
import numpy as np
```

```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

### The side-by-side histogram:

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5))

Air2008.ArrDelay.plot(kind = 'hist', title = 'Arrival Delay (in minutes)', 
                    color = 'purple', bins = 150, ax = ax1, 
                 xlim = (-65, 430), ylim = (-60, 5000))

Air2008.DepDelay.plot(kind = 'hist', title = 'Departure Delay (in minutes)',
                    color = 'purple', bins = 150, ax = ax2,
                 xlim = (-65, 430), ylim = (-60, 5000))

ax1.vlines(Air2008.ArrDelay.mean(), *ax1.get_ylim(), color='red', linewidth=3)
ax2.vlines(Air2008.DepDelay.mean(), *ax2.get_ylim(), color='red', linewidth=3)

ax1.vlines(Air2008.ArrDelay.median(), *ax1.get_ylim(), color='blue', linewidth=3)
ax2.vlines(Air2008.DepDelay.median(), *ax2.get_ylim(), color='blue', linewidth=3);
```

### The histogram with the density curve overlaying.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.distplot(Air2008.ArrDelay.dropna(), ax = ax1, hist = True)
sns.distplot(Air2008.DepDelay.dropna(), ax = ax2, hist = True)

ax1.set_title('Arrival Delay (in minutes)')
ax2.set_title('Departure Delay (in minutes)');
```

### The histogram with the vertical mean red line.

```python
ax = Air2008.ArrDelay.plot(kind='hist', bins=80, figsize=(10, 5), title = 'Arrival Delay (in minutes)')

ax.set_xlim(0, 350)
ax.set_ylim(0, 7000)

ax.vlines(Air2008.ArrDelay.mean(), *ax.get_ylim(), color='red', linewidth=3)
print('Home win percent:', (Air2008.ArrDelay > 0).mean());
```
