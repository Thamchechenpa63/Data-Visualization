### The pie plot:

The sectors of the pie plot correspond to the categories and the size of the wedges correspond to the counts of each category.

```python
import os

import pandas as pd
import numpy as np
```

```python
print(os.getcwd())
```

```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

### The pie plot:

The sectors of the pie plot correspond to the day of the week and the size of the wedges correspond to the counts.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Air2008.DayOfWeek.value_counts().plot(kind = 'pie', subplots = True, legend = False);
```

### The side-by-side pieplots:

```python
df = Air2008.groupby('Origin').ArrDelay.agg(['mean', 'median', 'std']).head()
df.plot(kind = 'pie', subplots = True, legend = False);
```
