```python
Air2008 = Airlines.sample(frac = 0.001, random_state = 90)
```

```python
% matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(Air2008, x_vars = ['WeatherDelay', 'DepDelay', 'ArrDelay'], 
             y_vars = 'ArrDelay', size = 8, aspect = 0.7, kind = 'reg');
```

### The scatterplot matrix:

1. The scatterplot matrix represents the correlation of pairs of the multiple features. However, the diagonal of the scatterplot matrix is exceptional. It would be pointless to correlate a factor to itself, so the space is used to display histograms. Scatterplots that are mirror images across the diagonal are reflections of each other, because they correlate the same two factors in the opposite order. 

2. The correlation coeffient r describes the strenght and the direction of the association between two quantitative features.

```python
sns.pairplot(Air2008, vars = ['ArrDelay', 'WeatherDelay', 'DepDelay'],
                 kind = 'reg', markers = '.');
```

It is possible to introduce colors to represent information about an additional feature. In this case, we are introducing the feature, day of the week (DayOfWeek). 

```python
sns.pairplot(Air2008, vars = ['ArrDelay', 'WeatherDelay', 'DepDelay'],
                 kind = 'reg', hue = 'DayOfWeek', markers = '.');
```
