### The violinplot:

A violin plot plays a similar role as a box and whisker plot. The violin plot also features a kernel density estimation of the underlying distribution.

```python
import os

import pandas as pd
import numpy as np
```

```python
print(os.getcwd())
```

```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

### The side-by-side violinplots:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.violinplot(Air2008.ArrDelay, ax = ax1, orient = 'v', color = 'red')
sns.violinplot(Air2008.DepDelay, ax = ax2, orient = 'v', color = 'green')

ax1.set_title('Arrival Delay (in minutes)')
ax2.set_title('Departure Delay (in minutes)');
```

### The horizontal violinplots:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.violinplot(Air2008.ArrDelay, ax = ax1, orient = 'h', color = 'blue')
sns.violinplot(Air2008.DepDelay, ax = ax2, orient = 'h', color = 'lightblue')

ax1.set_title('Arrival Delay (in minutes)')
ax2.set_title('Departure Delay (in minutes)');
```

### The multiple violinplots:

```python
ax = sns.violinplot(x = "DayOfWeek", y = "ArrDelay", data = Air2008)
ax.set_title('The Violin Plot');
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

### The multiple violinplots:

```python
ax = sns.violinplot(x = "survived", y = "fare", hue = "sex", data = titanic, palette="muted")
ax.set_title('The Violin Plot');
```

### The multiple violinplots:

```python
ax = sns.violinplot(x = "survived", y = "fare", hue = "sex", data = titanic, palette="muted", split=True)
```

### The side-by-side violinplots:

```python
ax = sns.violinplot(x = "sex", y = "survived", data = titanic, order=["male", "female"])
ax.set_title('The Violin Plot');
```

### The side-by-side violinplots:

```python
ax = sns.violinplot(x = "sex", y = "survived", data = titanic, 
                    palette = "Set3", split = True, scale = "width", inner = 'stick')
ax.set_title("The Violin Plot");
```

### The side-by-side violinplots:

```python
ax = sns.violinplot(x = "sex", y = "pclass", hue = "survived", data = titanic, 
                    palette = "Set2", split = True, scale = "count", inner = 'stick')
ax.set_title("The Violin Plot");
```

### The side-by-side violinplots:

```python
ax = sns.violinplot(x = "sex", y = "pclass", hue = "survived", data = titanic, 
                    palette = "Set2", split = True, scale = "count", inner = 'quartile')
ax.set_title("The Violin Plot");
```
