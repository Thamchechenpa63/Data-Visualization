### The bargraphs:

Bargaphs are plotted using categorical variables. The counts of the categorical variables are shown on the vertical axis. Categories are shown on the horizontal axis.

```python
import os

import pandas as pd
import numpy as np
```

```python
print(os.getcwd())
```

```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

### The vertical bar plots:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

Air2008.Origin.value_counts().head(10).plot(kind = 'bar', title = 'Flight Origin', color = 'lightgreen',ax = ax1)
Air2008.Dest.value_counts().head(10).plot(kind = 'bar', title = 'Flight Destination', color = 'lightblue', ax = ax2);
```

### The horizontal bar plots:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

Air2008.Origin.value_counts().head(10).sort_values().plot.barh(ax = ax1, color = 'lightblue')
Air2008.Dest.value_counts().head(10).sort_values().plot.barh(ax = ax2, color = 'lavender');
```

### The side-by-side barplot & the stacked barplot.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

Air2008.groupby('Origin').ArrDelay.agg(['mean', 'median', 'std']).head().plot(kind = 'bar',  
                                                                              ax = ax1)
Air2008.groupby('Origin').ArrDelay.agg(['mean', 'median', 'std']).head().plot(kind = 'bar', 
                                                                              stacked = True, ax = ax2);
```
