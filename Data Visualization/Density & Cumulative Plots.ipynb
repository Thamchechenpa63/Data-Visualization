### The kernel density estimation and cumulative distribution plots:

```python
import os

import pandas as pd
import numpy as np
```

```python
print(os.getcwd())
```

```python
Airlines = pd.read_csv('Air2008_copy.csv')
```

```python
Airlines.info(memory_usage = 'deep')
```

```python
Air2008 = Airlines.sample(frac = 0.01, random_state = 99)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.distplot(Air2008.ArrDelay.dropna(), ax = ax1, hist = False)
sns.distplot(Air2008.DepDelay.dropna(), ax = ax2, hist = False)

ax1.set_title('Kernel Density Estimation Plot of Arrival Delay')
ax2.set_title("Kernel Density Estimation Plot of Departure Delay");
```

```python
df = Air2008[['ArrDelay', 'DepDelay']]

fig, ax = plt.subplots()
df.plot.kde(ax = ax)

ax.set_title('Kernel Density Estimation Plot');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.kdeplot(Air2008.ArrDelay, shade = True, ax = ax1, cumulative = True)
sns.kdeplot(Air2008.DepDelay, shade = True, ax = ax2, cumulative = True)

ax1.set_xlim(-100, 1500)

ax1.set_title('The Cumulative Plot')
ax2.set_title('The Cumulative Plot');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

sns.kdeplot(Air2008.ArrDelay, shade = False, ax = ax1, cumulative = True)
sns.kdeplot(Air2008.DepDelay, shade = False, ax = ax2, cumulative = True)

ax1.set_xlim(-100, 1500)

ax1.set_title('The Cumulative Plot')
ax2.set_title('The Cumulative Plot');
```
